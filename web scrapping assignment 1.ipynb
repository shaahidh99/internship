### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make a data frame


```python
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

```python
movie_name=[]
year=[]
rating=[]
```

```python
pages = np.arange(1,100,50)
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text, 'html.parser')
    movie_data = soup.find_all('div', attrs = {'class': 'lister-item mode-advanced'})
    for store in movie_data:
        name = store.h3.a.text
        movie_name.append(name)
        
        year_of_release= store.h3.find('span', class_ = "lister-item-year text-muted unbold").text
        year.append(year_of_release)
        
        rate = store.find('div', class_ = "inline-block ratings-imdb-rating").text.replace('\n', '')
        rating.append(rate)       
#creating a dataframe 
movie_list = pd.DataFrame({ "Movie Name": movie_name, "Year of Release" : year,"Movie Rating": rating})
```

## 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page= requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
content= BeautifulSoup(page.content)

#collecting the data
title=[]
rating=[]
year=[]
for movie_name in content.find_all('td',class_='titleColumn'):
    c= movie_name.find('a')
    title.append(c.text)
    
for movie_year in content.find_all('td',class_='titleColumn'):
    c= movie_year.find('span')
    year.append(c.text)
    
for movie_rating in content.find_all('td',class_='ratingColumn imdbRating'):
    rating.append(movie_rating.text[1])
    
    
#creating Dataframe
top_rated_indian_movies= pd.DataFrame({"Title": title,'Rating':rating,'Year':year})
full_list=top_rated_indian_movies[0:101]
   
```

## Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp

```python
#requesting and parsing the content of the page
page= requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
content= BeautifulSoup(page.content)

#creating list for collecting the required data
product=[]
price=[]
discounts=[]

#collecting the data (names, rate, discounts)
for i in content.find_all('div',class_='sc-dkPtRN ProductList__GridCol-sc-8lnc8o-0 FjWWx jMkQHN'):
    names= i.find('p')
    rate= i.find('h5',class_='Text__StyledText-sc-oo0kvp-0 hiHdyy')
    discount=i.find('p',class_='Text__StyledText-sc-oo0kvp-0 fCJVtz NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm NewProductCard__DiscountTextParagraph-sc-j0e7tu-16 kmYsnm')
    product.append(names.text)
    price.append(rate.text.replace('₹',''))
    discount=discount.text.split(' ')
    discounts.append(discount[0].replace('₹',''))
    
#creating Dataframe
Handbag_price= pd.DataFrame({"Product": product,'Discount on first order':discounts,'Price':price})
Handbag_price
        
```
