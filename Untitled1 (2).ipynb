```python
from selenium import webdriver
import os
import time
import urllib.request
from selenium.webdriver.common.keys import Keys


driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.google.com/')
search = driver.find_element_by_name('q')
search.send_keys('people wearing mask',Keys.ENTER)


#creating a directory to save images
folder_name = 'images'
if not os.path.isdir(folder_name):
    os.makedirs(folder_name)

    
elem = driver.find_element_by_link_text('Images')
elem.get_attribute('href')
elem.click()
value = 0
for i in range(10):
	driver.execute_script('scrollBy("+ str(value) +",+500);')
	value += 100
	time.sleep(4)
elements = driver.find_elements_by_xpath('//img[contains(@class,"rg_i")]')
count = 0
for i in elements:
    src = i.get_attribute('src')
    try:
        if src != None:
            src  = str(src)
            count+=1
            urllib.request.urlretrieve(src, os.path.join(folder_name,'with_mask_image'+str(count)+'.jpg'))
            if count%10 == 0: print("downloaded",count,"images")
        else:
            raise TypeError
    except TypeError:
        pass
```

```python
driver= webdriver.Chrome('chromedriver.exe')
url='https://www.flipkart.com/'
driver.get(url)

product_name= input('Enter the product you want to search ')
search= driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")
search.send_keys(product_name)

search_btn= driver.find_element_by_xpath("/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search_btn.click()
```

```python
name=driver.find_elements_by_xpath('//div[@class="_4rR01T"]')
product=[]
for i in name:
    if i.text=='':
        continue
    else:
        product.append(i.text)
        
brand= []
for i in product:
    brand.append((i.split()[0]))
    
phone=[]
color=[]

for i in product:
    i=i.split('(')
    phone.append((i[0]))
    color.append((i[1].split(',')[0]))


details= driver.find_elements_by_xpath('//ul[@class="_1xgFaf"]')
detail=[]
for i in details:
    if i.text=='':
        continue
    else:
        detail.append(i.text)
        
rom=[]
ram=[]
display=[]
frontcam=[]
rearcam=[]
bat=[]
for i in detail:
    i=i.split('|')
    ram.append((i[0]))
    j=i[1].split("\n")
    rom.append((i[1].split("\n"))[0])
    if 'cm' in i[1]:
        display.append(j[1])
    else:
        display.append('-')
    if 'MP' in i[1]:
        rearcam.append(j[2])
    else:
        rearcam.append('-')
    j=i[2].split("\n")
    if 'Front' in i[2]:
        frontcam.append(j[0])
    else:
        frontcam.append('-')
    if 'Battery' in i[2]:
        bat.append(j[1])
    else:
        bat.append('-')
        
pricing= driver.find_elements_by_xpath('//div[@class="_30jeq3 _1_WHN1"]')
price=[]
for i in pricing:
    price.append(i.text)
    
html= driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')
link=[]
for i in html:
    link.append(i.get_attribute('href'))
    
    
smartphone= pd.DataFrame({'Brand':brand,'Smartphone':phone,'color':color,'ROM':rom,'RAM':ram,
                         'Display':display,'Front Cam':frontcam,'Rear Cam':rearcam,
                         'Battery':bat,'Price':price,'Link':link})
smartphone
```

```python
driver= webdriver.Chrome('chromedriver.exe')

url='https://trak.in/'
driver.get(url)


fund= driver.find_element_by_xpath('/html/body/div[1]/header/div[2]/div/div/div/div/nav/ul/li[9]/a')
fund.click()

date= driver.find_elements_by_xpath('//td[@class="column-2"]')
startup= driver.find_elements_by_xpath('//td[@class="column-3"]')
Industry_Vertical= driver.find_elements_by_xpath('//td[@class="column-4"]')
SubVertical= driver.find_elements_by_xpath('//td[@class="column-5"]')
City_Location= driver.find_elements_by_xpath('//td[@class="column-6"]')
Investors_Name= driver.find_elements_by_xpath('//td[@class="column-7"]')
Investors_type= driver.find_elements_by_xpath('//td[@class="column-8"]')
Amount_In_USD= driver.find_elements_by_xpath('//td[@class="column-9"]')


start= [i.text for i in startup]
Industry=[i.text for i in Industry_Vertical]
sub= [i.text for i in SubVertical]
loc= [i.text for i in City_Location]
inv_name= [i.text for i in Investors_Name]
inv_type=[i.text for i in Investors_type]
amount=[i.text for i in Amount_In_USD]


dates=[]
for i in date:
    dates.append(i.text)
```

```python
funds= pd.DataFrame()
funds['Date']=dates[:392]
funds['Startup Name']= start
funds['Industry / Vertical']= Industry
funds['Sub-Vertical']= sub
funds['City / Location']=loc
funds['Investors Name']=inv_name
funds['Investment Type']= inv_type
funds['Amount (In USD)']=amount


funds.drop(index=(funds.index[funds.Date=='05/072018']),inplace=True)
funds['Date']=pd.to_datetime(funds.Date)
mask = (funds['Date'] > '01/01/2021') & (funds['Date'] <= '31/03/2021')
funds_2nd_quaterly=funds[mask].reset_index(drop=True)
funds_2nd_quaterly
```

```python
name=[]
name_tag= driver.find_elements_by_xpath('//*[@class="TopNumbeHeading sticky-footer"]/a/h3')
for i in name_tag:
    if i.text=='':
        continue
    else:
        name.append(i.text)
        
specs=[]
specs_tag= driver.find_elements_by_xpath('//div[@class="Spcs-details"]')
for i in specs_tag:
    specs.append(i.text)
    
    
os=[]
display=[]
processor=[]
memory=[]
weight=[]
dimension=[]
graphics=[]
for i in specs:
    i=i.split(":")
    os.append((i[1]).split('\n')[0])
    display.append((i[2]).split('\n')[0])
    processor.append((i[3]).split('\n')[0])
    memory.append((i[4]).split('\n')[0])
    weight.append((i[5]).split('\n')[0])
    dimension.append((i[6]).split('\n')[0])
    graphics.append((i[7]).split('\n')[0])
    
    
product= pd.DataFrame({'Name':name,'OS':os[:9],'Display':display[:9],'Processor':processor[:9],
                      'memory':memory[:9],'weight':weight[:9],'dimension':dimension[:9],'graphics':graphics[:9]})
product 
```
